```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
with open(r'C:\Users\imdan\dscamp\DSCapstoneProj\data\external\USCODE22_LLCP_102523.HTML') as file:
    html_content = file.read()
    
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
# we want to find variable name and question?
#so what do i want to do? i want to parse each table and extract the code/var name and the question, tuple key:value
```

```python
#seperates on tag 'td'(table cell), on attribute where class = 'l m linecontent''
results = soup.find_all('td', attrs={'class':'l m linecontent'})
results
```

```python
test=results[1].contents
print(test)
```

```python
list = []

for elements in results:
    items = elements.contents
    temp_tup = []
    for item in items:
        #need to change to string because of bs4 element tag.
        string = str(item)
        if string.startswith('SAS'):
            temp_tup.append(string.replace('\xa0','').split(':')[1])
        if string.startswith('Question:'):
            temp_tup.append(string.replace('\xa0',' ').split(':')[1].strip())
    list.append(temp_tup)

Data_Dict = pd.DataFrame(list)
Data_Dict.columns = ['VAR_NAME','QUESTION']
Data_Dict
```
