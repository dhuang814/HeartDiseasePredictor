Mapping and transforming data for readability and allows us to analyze the data easier

```python
pass3['State'] = pass3['State'].map(state)
pass3['Sex'] = pass3['Sex'].map(sex)
pass3['GeneralHealth'] = pass3['GeneralHealth'].map(generalhealth)
pass3['PoorPhysicalHealthDays'] = pass3['PoorPhysicalHealthDays'].replace(healthdays)
pass3['PoorMentalHealthDays'] = pass3['PoorMentalHealthDays'].replace(healthdays)
pass3['LastCheckup'] = pass3['LastCheckup'].map(checkup)
pass3['SleepTime'] = pass3['SleepTime'].apply(sleeptime)
pass3['HadCovid'] = pass3['HadCovid'].map(covid)
pass3['HeartDisease'] = pass3['HeartDisease'].map(yes_no)
pass3['RaceEthnicityGroup'] = pass3['RaceEthnicityGroup'].map(race)
pass3['TestedForHIV'] = pass3['TestedForHIV'].map(yes_no)
pass3['AgeGroup5yrs'] = pass3['AgeGroup5yrs'].map(agegroup)

#broke
pass3['LevelOfSmoker'] = pass3['LevelOfSmoker'].map(smoke)
pass3['HadTetanusShot'] = pass3['HadTetanusShot'].map(tetanus)
pass3['EcigUsage'] = pass3['EcigUsage'].map(ecig)
pass3['HadDiabetes'] = pass3['HadDiabetes'].map(diabetes)
```

```python
#if nunique is 4 then yes_no 
exclude = ['LevelOfSmoker','HadTetanusShot','EcigUsage','HadDiabetes']

for i in pass3:
    if pass3[i].nunique() == 4 and i not in exclude:
        pass3[i] = pass3[i].map(yes_no)
```

```python
pass3['DrinkOccasionsPerDay'] = pass3['DrinkOccasionsPerDay'].astype(int)
pass3['DrinkOccasionsPerDay'] = pass3['DrinkOccasionsPerDay'].replace(alc)
```
